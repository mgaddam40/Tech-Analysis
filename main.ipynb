Database of 500 Students' answers

```python
import random
import csv

num_arrays = 500
array_size = 20

# Generate student numbers
start_roll_num = 22001
end_roll_num = 22500
roll_nums = list(range(start_roll_num, end_roll_num + 1))

# Generate 500 arrays with student numbers
arrays = []
for roll_num in roll_nums:
    new_array = [roll_num] + [random.choice([-1, 0, 1]) for _ in range(array_size)]  # Randomly select from -1, 0, 1
    arrays.append(new_array)

# Write the arrays to a CSV file
csv_filename = "exam_data.csv"
with open(csv_filename, "w", newline="") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["roll_num"] + [f"Q{i + 1}" for i in range(array_size)])  # Write header row

    for array in arrays:
        csv_writer.writerow(array)

print(f"CSV file '{csv_filename}' generated successfully.")
```

```python
# Storing the data from QP_data into arrays
import csv

# Initialize lists to store the data
ques_nums = []
difficulty = []
deduction_or_reasoning = []
numerical_logic = []
logical_intuition = []
trial_and_error = []
visualisation = []
total_scores = []

# Read the CSV file
csv_file_path = 'QP_data.csv'  # Update this with the correct file path
with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        ques_nums.append(int(row['que']))
        difficulty.append(int(row['difficulty']))
        deduction_or_reasoning.append(int(row['deduction_or_reasoning']))
        numerical_logic.append(int(row['numerical_logic']))
        logical_intuition.append(int(row['logical_intuition']))
        trial_and_error.append(int(row['trial_and_error']))
        visualisation.append(int(row['visualisation']))
        total_scores.append(int(row['total']))

# Print the arrays to verify the data
print("Question Numbers:", ques_nums)
print("Difficulty:", difficulty)
print("Deduction or Reasoning:", deduction_or_reasoning)
print("Numerical Logic:", numerical_logic)
print("Logical Intuition:", logical_intuition)
print("Trial and Error:", trial_and_error)
print("Visualisation:", visualisation)
print("Total Scores:", total_scores)
```

```python
# Performace of students
roll_num = []
stu_difficulty = []
stu_deduction_or_reasoning = []
stu_numerical_logic = []
stu_logical_intuition = []
stu_trial_and_error = []
stu_visualisation = []

# Total values of performance
tot_difficulty = sum(difficulty)
tot_deduction_or_reasoning = sum(deduction_or_reasoning)
tot_numerical_logic = sum(numerical_logic)
tot_logical_intuition = sum(logical_intuition)
tot_trial_and_error = sum(trial_and_error)
tot_visualisation = sum(visualisation)

j = 0
for student in arrays:
    corr_difficulty_sum = 0
    corr_deduction_or_reasoning_sum = 0
    corr_numerical_logic_sum = 0
    corr_logical_intuition_sum = 0
    corr_trial_and_error_sum = 0
    corr_visualisation_sum = 0

    for i in range(1,len(student)):
        if student[i] == 1:
            corr_difficulty_sum += difficulty[i-1]
            corr_deduction_or_reasoning_sum += deduction_or_reasoning[i-1]
            corr_numerical_logic_sum += numerical_logic[i-1]
            corr_logical_intuition_sum += logical_intuition[i-1]
            corr_trial_and_error_sum += trial_and_error[i-1]
            corr_visualisation_sum += visualisation[i-1]

    roll_num.append(arrays[j][0])
    j += 1
    stu_difficulty.append(5 * corr_difficulty_sum / tot_difficulty)
    stu_deduction_or_reasoning.append(5 * corr_deduction_or_reasoning_sum / tot_deduction_or_reasoning)
    stu_numerical_logic.append(5 * corr_numerical_logic_sum / tot_numerical_logic)
    stu_logical_intuition.append(5 * corr_logical_intuition_sum / tot_logical_intuition)
    stu_trial_and_error.append(5 * corr_trial_and_error_sum / tot_trial_and_error)
    stu_visualisation.append(5 * corr_visualisation_sum / tot_visualisation)

print(roll_num)
print(stu_difficulty)
print(stu_deduction_or_reasoning)
print(stu_numerical_logic)
print(stu_logical_intuition)
print(stu_trial_and_error)
print(stu_visualisation)
```

Creating a csv sheet with the above generated data

```python
data = list(zip(roll_num, stu_difficulty, stu_deduction_or_reasoning, stu_numerical_logic, stu_logical_intuition, stu_trial_and_error, stu_visualisation))

# Specify the CSV file path
csv_file_path = 'student_performance_data.csv'

# Write the data to the CSV file
with open(csv_file_path, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    
    # Write column names
    column_names = ['roll_num', 'difficulty', 'deduction_or_reasoning', 'numerical_logic', 'logical_intuition', 'trial_and_error', 'visualisation']
    csv_writer.writerow(column_names)
    
    # Write data rows
    csv_writer.writerows(data)

print(f"Data has been written to {csv_file_path}")
```
